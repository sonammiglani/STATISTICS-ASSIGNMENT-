# Q1. What is a random variable in probability theory?
A **random variable** in probability theory is a function that assigns numerical values to the outcomes of a random experiment. It helps quantify uncertainty and allows mathematical analysis of probabilistic events.

Random variables are classified into two main types:

# Q6.What is a discrete uniform distribution?
A **discrete uniform distribution** is a probability distribution where a finite number of **equally likely** outcomes are possible. That means **each outcome has the same probability** of occurring.


### Example:

If you roll a fair six-sided die, the outcomes are:

$$
1, 2, 3, 4, 5, 6
$$

Each has a probability of:

$$
P(X = x) = \frac{1}{6}
$$

for $x = 1, 2, 3, 4, 5, 6$.

### General Formula:

If there are $n$ possible outcomes $x_1, x_2, ..., x_n$, then:

$$
P(X = x_i) = \frac{1}{n} \quad \text{for all } i
$$

# Q7. What are the key properties of a Bernoulli distribution?
The **Bernoulli distribution** is a discrete probability distribution for a random variable which has exactly two possible outcomes: success (usually coded as 1) and failure (usually coded as 0). Here are its key properties:

 1. **Outcomes**

* Only two outcomes: 0 and 1.

 2. **Probability Parameter (p)**

* $p = P(X = 1)$, the probability of success.
* $1 - p = P(X = 0)$, the probability of failure.
* $0 \leq p \leq 1$

3. **Probability Mass Function (PMF)**

$$
P(X = x) = p^x (1 - p)^{1 - x}, \quad x \in \{0, 1\}
$$

4. **Mean (Expected Value)**

$$
E[X] = p
$$

 5. **Variance**

$$
Var(X) = p(1 - p)
$$

 6. **Support**

* $x \in \{0, 1\}$

 7. **Moment Generating Function (MGF)**

$$
M_X(t) = (1 - p) + pe^t
$$

 8. **Skewness**

$$
\frac{1 - 2p}{\sqrt{p(1 - p)}}
$$

 9. **Kurtosis**

$$
\frac{1 - 6p(1 - p)}{p(1 - p)}
$$

The Bernoulli distribution is the foundation of the **Binomial distribution**, which models the number of successes in a fixed number of independent Bernoulli trials.

# Q8. What is the binomial distribution, and how is it used in probability?
The **binomial distribution** is a probability distribution that summarizes the likelihood of a **fixed number of successes** in a **fixed number of independent trials**, where each trial has only two possible outcomes: **success** or **failure**. It's used when you're counting how often a particular outcome occurs across several repeated experiments.


Probability Mass Function (PMF):

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

Where:

* $\binom{n}{k}$ is the number of combinations (n choose k)
* $k$ is the number of successes
* $n$ is the total number of trials
* $p$ is the probability of success

The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space, given that the events happen with a known constant mean rate and independently of the time since the last event.

Common Application--

Traffic engineering: Cars passing a checkpoint in an hour.

Biology: Number of mutations in a DNA strand over a length.

Finance: Number of trades per second.

Astronomy: Number of photons hitting a detector in a fixed time.

Retail: Number of customers arriving at a store in a given period.

# Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes how the distribution of sample means approaches a normal distribution, as the sample size increases, regardless of the original distribution of the population.

**CRITICAL**

The **Central Limit Theorem (CLT)** is critical because it enables statisticians to use normal distribution-based methods (like hypothesis testing and confidence intervals) even when the underlying population is not normal, as long as the sample size is large enough. This makes statistical analysis more versatile and applicable in real-world situations.

# Q15.What is the application of Z statistics in hypothesis testing?


1. **Testing Population Means** (when population standard deviation is known):

   * Used to compare sample mean to the population mean.
   * Formula:

     $$
     Z = \frac{\bar{x} - \mu_0}{\frac{\sigma}{\sqrt{n}}}
     $$

     where $\bar{x}$ = sample mean, $\mu_0$ = hypothesized population mean, $\sigma$ = population standard deviation, and $n$ = sample size.

2. **Testing Population Proportions**:

   * Used to test if the sample proportion is different from the hypothesized population proportion.
   * Formula:

     $$
     Z = \frac{p_{\text{hat}} - p_0}{\sqrt{\frac{p_0(1 - p_0)}{n}}}
     $$

     where $p_{\text{hat}}$ = sample proportion, $p_0$ = hypothesized population proportion, and $n$ = sample size.

3. **Large Sample Sizes**:

   * Z-tests are typically used for large samples (n > 30), as the sampling distribution of the sample mean approaches normality due to the Central Limit Theorem.

4. **Standard Normal Distribution**:

   * Z-scores follow the standard normal distribution, and the Z-statistic is compared to critical values from this distribution to make decisions.

5. **Comparison to Critical Values**:

   * The calculated Z-value is compared to a critical value (e.g., $\pm 1.96$ for 95% confidence level) to determine whether to reject the null hypothesis.

6. **Example in Practice**:

   * Used to test if a company's claim (e.g., average lifespan of light bulbs) is statistically different from sample data using known population standard deviation.

# Q16. How do you calculate a Z-score, and what does it represent?
A Z-score, also called a standard score, measures how far a data point is from the mean of a dataset in terms of standard deviations. It helps determine whether a value is typical or unusual within a given distribution.
Formula for Calculating a Z-score
[ Z = \frac{X - \mu}{\sigma} ] Where:
- (X) = The individual data point
- (\mu) = The mean of the dataset
- (\sigma) = The standard deviation of the dataset
What a Z-score Represents
- A Z-score of 0 means the data point is exactly at the mean.
- A positive Z-score (e.g., +1.5) indicates the data point is above the mean.
- A negative Z-score (e.g., -2.0) suggests the data point is below the mean.
- A larger absolute Z-score (e.g., Â±3.0 or beyond) often signifies an outlier or an unusually rare event in a normal distribution.

# Q22.  What is the concept of expected value in a probability distribution?
The **expected value** in a probability distribution is the **long-run average outcome** you would expect if you repeated an experiment many times.



 **Definition:**

For a **discrete** random variable $X$, the expected value $E(X)$ is:

$$
E(X) = \sum [x_i \cdot P(x_i)]
$$

Where:

* $x_i$ = possible outcome
* $P(x_i)$ = probability of outcome $x_i$

For a **continuous** variable, it's calculated using an integral:

$$
E(X) = \int_{-\infty}^{\infty} x \cdot f(x)\, dx
$$



###  **Example (Discrete):**

Imagine a dice game:

* Roll a 6 â†’ win \$10
* Roll 1â€“5 â†’ win \$0

Then:

$$
E(X) = (10 \cdot \tfrac{1}{6}) + (0 \cdot \tfrac{5}{6}) = \tfrac{10}{6} \approx 1.67
$$

So, **on average**, win OF \$1.67 per roll over many trials.

# Q23.  How does a probability distribution relate to the expected outcome of a random variable?
A probability distribution describes how the values of a random variable are spread out, assigning probabilities to each possible outcome of that variable. It provides the likelihood of different outcomes occurring in a random experiment.

The **expected outcome** (or expected value) of a random variable is the weighted average of all possible outcomes, with the probabilities acting as the weights. This gives a sense of the "center" or "average" value that the random variable is likely to take over many trials.

Mathematically, for a discrete random variable $X$ with possible outcomes $x_1, x_2, \dots, x_n$ and corresponding probabilities $p_1, p_2, \dots, p_n$, the expected value $E(X)$ is:

$$
E(X) = \sum_{i=1}^{n} x_i p_i
$$

For a continuous random variable, the expected value is calculated as the integral of the product of the variable's value and its probability density function $f(x)$:

$$
E(X) = \int_{-\infty}^{\infty} x f(x) dx
$$
